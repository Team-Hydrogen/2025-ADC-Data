```python
import pandas as pd
import matplotlib.pyplot as plt

nominaldf = pd.read_csv('./data/starting_data/hsdata.csv')
offdf = pd.read_csv('./data/starting_data/bonusdata.csv')
```

```python
offdf["DX (km)"] = offdf["Rx(km)[J2000-EARTH]"]-nominaldf["Rx(km)[J2000-EARTH]"]
offdf["DY (km)"] = offdf["Ry(km)[J2000-EARTH]"]-nominaldf["Ry(km)[J2000-EARTH]"]
offdf["DZ (km)"] = offdf["Rz(km)[J2000-EARTH]"]-nominaldf["Rz(km)[J2000-EARTH]"]
offdf.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

offdf.fillna(0, inplace=True)

X=offdf[["Rx(km)[J2000-EARTH]","Ry(km)[J2000-EARTH]","Rz(km)[J2000-EARTH]","Vx(km/s)[J2000-EARTH]","Vy(km/s)[J2000-EARTH]","Vz(km/s)[J2000-EARTH]"]]
Y=offdf[["DX (km)","DY (km)","DZ (km)"]]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1)

model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
time = 60

thrust=[]
for i in range(offdf.shape[0]):
    x = offdf.iloc[i][["Rx(km)[J2000-EARTH]","Ry(km)[J2000-EARTH]","Rz(km)[J2000-EARTH]","Vx(km/s)[J2000-EARTH]","Vy(km/s)[J2000-EARTH]","Vz(km/s)[J2000-EARTH]"]].values
    mass = offdf.iloc[i]["MASS (kg)"]
    dx,dy,dz = model.predict([x])[0]
    thrust.append([mass*dx/(time**2),mass*dy/(time**2),mass*dz/(time**2)])

offdf["ThrustX (N)"] = [x[0] for x in thrust]
offdf["ThrustY (N)"] = [x[1] for x in thrust]
offdf["ThrustZ (N)"] = [x[2] for x in thrust]
offdf.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(offdf["ThrustX (N)"], offdf["ThrustY (N)"], offdf["ThrustZ (N)"], label='Thrust Vector', color='blue')
ax.set_xlabel('Thrust X (N)')
ax.set_ylabel('Thurst Y (N)')
ax.set_zlabel('Thrust Z (N)')
plt.legend()
plt.show()
```
